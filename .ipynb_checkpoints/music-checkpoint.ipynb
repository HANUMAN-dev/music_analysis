```python
query3 = pd.read_sql("""
SELECT COUNT(T.album_id) AS C,
    A.title FROM album A
   JOIN track  T ON A.album_id=T.album_id

 GROUP BY T.album_id
 ORDER BY  C DESC LIMIT 1
;
""", conn)

query3
```

Which artist or album has the highest number of tracks in the store?

```python
query3 = pd.read_sql("""
SELECT  A.artist_id,COUNT(A.artist_id) AS Count_in_store
     FROM album A
   JOIN track  T ON A.album_id=T.album_id
   GROUP BY
    A.artist_id
    ORDER BY Count_in_store DESC LIMIT 1
    
    
;










""", conn)

query3

```

Determine the number of tracks for each album and calculate the total duration of each album

```python
query3 = pd.read_sql("""
SELECT COUNT(T.track_id) AS TrackCount,
    A.title,sum(T.milliseconds) AS duration FROM album A
   JOIN track  T ON A.album_id=T.album_id

 GROUP BY T.album_id
 ORDER BY TrackCount DESC
 
;
""", conn)

query3
```

Identify the albums that have the highest average track duration

```python
query3 = pd.read_sql("""
SELECT  
    A.title AS Title,avg(T.milliseconds) AS duration FROM album A
   JOIN track  T ON A.album_id=T.album_id

 GROUP BY T.album_id
 ORDER BY duration DESC
 
;
""", conn)

query3
```

How does the genre distribution vary among albums

```python
query3 = pd.read_sql("""
SELECT  
    T.name AS Title, COUNT(A.genre_id) AS Genrecount FROM genre A
   JOIN track  T ON A.genre_id=T.genre_id

   GROUP BY T.album_id ORDER BY  Genrecount
;
""", conn)

query3
```

Identify the average number of tracks purchased per customer for each genre.

```python
query1 = pd.read_sql("""
SELECT count(D.track_id),A.first_name
FROM customer  A JOIN invoice B ON A.customer_id=B.customer_id
JOIN invoice_line C ON B.invoice_id=C.invoice_id
JOIN track D ON D.track_id=C.track_id
group by D.genre_id,D.genre_id,A.customer_id
;
""", conn)

query1
```

What is the average purchase value per customer?

```python
query1 = pd.read_sql("""
SELECT A.first_name,avg(B.total)
FROM customer  A JOIN invoice B ON A.customer_id=B.customer_id
JOIN invoice_line C ON B.invoice_id=C.invoice_id
group by A.customer_id
order by avg(B.total) DESC limit 10
;
""", conn)

query1
```

Which customers have made the highest number of purchases?

```python
query1 = pd.read_sql("""
SELECT A.first_name,COUNT(B.total) AS Number_of_purchase
FROM customer  A JOIN invoice B ON A.customer_id=B.customer_id
JOIN invoice_line C ON B.invoice_id=C.invoice_id
group by A.customer_id
order by Number_of_purchase desc limit 1
;
""", conn)

query1
```

Calculate the percentage of total sales contributed by each genre in the USA and identify the best-selling genres and artists.

```python
query1 = pd.read_sql("""
SELECT   E.name,         (  TOTAL/ (SELECT SUM(total) FROM (
SELECT sum(A.total) AS TOTAL, D.name,billing_country,C.album_id
FROM invoice  A JOIN invoice_line B ON A.invoice_id=B.invoice_id
JOIN track C ON C.track_id=B.track_id
JOIN genre D ON D.genre_id=C.genre_id
WHERE A.billing_country='USA'
GROUP BY C.genre_id )))*100       AS sales_contribution FROM (
SELECT sum(A.total) AS TOTAL, D.name,billing_country,C.album_id,D.name
FROM invoice  A JOIN invoice_line B ON A.invoice_id=B.invoice_id
JOIN track C ON C.track_id=B.track_id
JOIN genre D ON D.genre_id=C.genre_id
WHERE A.billing_country='USA'
GROUP BY C.genre_id )

E JOIN album F ON   E.  album_id=F.album_id
JOIN artist G ON G.artist_id=F.artist_id

ORDER BY           sales_contribution DESC
;   
""", conn)

query1
```

Use subqueries to find the top-selling tracks in the USA and identify their associated genres.


```python
query1 = pd.read_sql("""
SELECT sum(I.total) as sells_all,T.genre_id,G.name
FROM invoice I JOIN invoice_line IL ON I.invoice_id=IL.invoice_id
JOIN track T on T.track_id=IL.track_id
JOIN genre G ON G.genre_id=T.genre_id
where I.billing_country='USA'
GROUP BY T.track_id 
ORDER BY sells_all desc
LIMIT 10
;
""", conn)

query1
```

Implement joins to retrieve customer information along with their preferred genres and country of origin.

```python
query1 = pd.read_sql("""
SELECT * FROM (
SELECT I.Billing_country as Country,G.name as genre_name,I.customer_id
FROM invoice I JOIN invoice_line IL ON I.invoice_id=IL.invoice_id
JOIN track T on T.track_id=IL.track_id
JOIN genre G ON G.genre_id=T.genre_id



) T1 JOIN customer C ON T1.customer_id=C.customer_id
GROUP BY C.customer_id
;
""", conn)

query1
```

```python
query1 = pd.read_sql("""
SELECT A.title,SUM(I.total) AS SELLS
FROM invoice I JOIN invoice_line IL ON I.invoice_id=IL.invoice_id
JOIN track T on T.track_id=IL.track_id
JOIN genre G ON G.genre_id=T.genre_id
JOIN album A ON A.album_id=T.album_id
WHERE I.billing_country='USA'
GROUP BY T.album_id
ORDER BY SELLS DESC
LIMIT 3

;
""", conn)

query1
```
